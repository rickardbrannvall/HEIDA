```python
import pandas as pd
```

```python
df = pd.read_csv("./data/framingham.csv")
print(df.shape)
df = df.dropna()
print(df.shape)
df = df.drop(columns=["education", "currentSmoker", "BPMeds", "diabetes", "diaBP", "BMI"])#.dropna()
print(df.shape)
df = df.dropna()
print(df.shape)

df.head()
```

```python
df.describe()
```

```python
grouped = df.groupby('TenYearCHD')
grouped.count()
#grouped.apply(lambda x: 1)
```

```python
#data = grouped.apply(lambda x: x.sample(grouped.size().min(), random_state=73).reset_index(drop=True))
data = grouped.apply(lambda x: x.sample(500, random_state=73).reset_index(drop=True))
#data
data.describe()
```

```python
ndata = data.reset_index(drop=True).drop(columns="TenYearCHD").apply(lambda x: (x-x.mean())/ x.std(), axis=0)
ndata["TenYearCHD"] = data["TenYearCHD"].reset_index(drop=True)
ndata = ndata.sample(len(ndata), random_state=42)
ndata
```

```python
ndata.to_csv(f"data/myTenYearCHD_n{len(ndata)}.csv")
```
